## About Dataset
The dataset this data contains more of 10, 000 rows and more than 10 columns which contains features of the car such as Engine Fuel Type, Engine HP, Transmission Type, highway MPG, city MPG and many more. <br>

Dataset Description<br>
Make - Car Make<br>
Model - Car Model<br>
Year - Car Year <br>
Engine Fuel Type - Premium unleaded, Regular unleaded, other. <br>
Engine HP - Engine Horse Power<br>
Engine Cylinders - Between 0 and 16.<br>
Trnsmission Type - automatic /manual <br>
Driven Wheels - front wheel drive/rear wheel drive <br>
Number of doors - between 2 and 4 <br>
Market Category -N/A, Crossover, Other <br>
Vehicle Size - Compact, Midsize, Other <br>
Vehicle Sytle -Type of vehicle e.g Sedan <br>
Highway MPG - Highway Miles per Gallon <br>
City MPG - City Miles per Gallon <br>
Popularity - Popularity <br>
MSRP - Manufacturer Suggested Retail Price <br>
